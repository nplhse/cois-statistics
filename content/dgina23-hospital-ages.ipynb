```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('data/dgina23_hospitals_age.csv')

# reshape the d dataframe suitable for statsmodels package
d_melt = pd.melt(df, id_vars=['age'], value_vars=['basic_m', 'basic_w'])
d_melt = d_melt.dropna(subset=['age'])
d_melt.columns = ['age', 'group', 'value']

d2_melt = pd.melt(df, id_vars=['age'], value_vars=['extended_m', 'extended_w'])
d2_melt = d2_melt.dropna(subset=['age'])
d2_melt.columns = ['age', 'group', 'value']

d3_melt = pd.melt(df, id_vars=['age'], value_vars=['full_m', 'full_w'])
d3_melt = d3_melt.dropna(subset=['age'])
d3_melt.columns = ['age', 'group', 'value']

fig, axes = plt.subplots(3, 1, figsize=(12, 10))
fig.suptitle('Ages and hospital tiers')
fig.tight_layout(pad=2.5)

sns.lineplot(x="age", y="value", hue="group", data=d_melt, ax=axes[0])
sns.lineplot(x="age", y="value", hue="group", data=d2_melt, ax=axes[1])
sns.lineplot(x="age", y="value", hue="group", data=d3_melt, ax=axes[2])

plt.show();
```

```python
df = pd.read_csv('data/dgina23_hospitals_age_groups.csv')

# reshape the d dataframe suitable for statsmodels package
d_melt = pd.melt(df, id_vars=['age_group'], value_vars=['basic_m', 'basic_w', 'extended_m', 'extended_w', 'full_m', 'full_w'])
d_melt = d_melt.dropna(subset=['age_group'])
d_melt.columns = ['age_group', 'group', 'value']

d2_melt = pd.melt(df, id_vars=['age_group'], value_vars=['basic_m', 'basic_w'])
d2_melt = d2_melt.dropna(subset=['age_group'])
d2_melt.columns = ['age_group', 'group', 'value']

d3_melt = pd.melt(df, id_vars=['age_group'], value_vars=['extended_m', 'extended_w'])
d3_melt = d3_melt.dropna(subset=['age_group'])
d3_melt.columns = ['age_group', 'group', 'value']

d4_melt = pd.melt(df, id_vars=['age_group'], value_vars=['full_m', 'full_w'])
d4_melt = d4_melt.dropna(subset=['age_group'])
d4_melt.columns = ['age_group', 'group', 'value']

fig, axes = plt.subplots(4, 1, figsize=(12, 10))
fig.suptitle('Age groups and hospital tiers')
fig.tight_layout(pad=2.5)

sns.barplot(x="age_group", y="value", hue="group", data=d_melt, ax=axes[0])
sns.barplot(x="age_group", y="value", hue="group", data=d2_melt, ax=axes[1])
sns.barplot(x="age_group", y="value", hue="group", data=d3_melt, ax=axes[2])
sns.barplot(x="age_group", y="value", hue="group", data=d3_melt, ax=axes[3])

plt.show();
```
