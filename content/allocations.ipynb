```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('data/allocations.csv')
df = df.set_index(['month'])

plt.figure(figsize=(10, 10))
sns.set(font_scale=1.5)

sns.heatmap(df, cmap='coolwarm_r', annot=True, fmt='.5g')

plt.title('Heatmap of all allocations by year& month')
plt.xlabel('Year')
plt.ylabel('Month')
```
