```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv('data/hospitals.csv')

df["size"] = df["size"].astype("category")
df["size"] = df["size"].cat.set_categories(['Small', 'Medium', 'Large'], ordered=True)
df["tier"] = df["tier"].astype("category")
df["tier"] = df["tier"].cat.set_categories(['Basic', 'Extended', 'Full'], ordered=True)
```

```python
plt.subplots(figsize = (10,8))
plt.tight_layout()

plt.subplot(1, 2, 1)
plt.title("Hospital size")
plt.pie(labels=['Small < 250 beds', 'Medium', 'Large > 750 beds'], x=df.value_counts('size', sort=False), autopct='%1.1f%%')

plt.subplot(1, 2, 2)
plt.title("Hospital tiers")
plt.pie(labels=['Basic', 'Extended', 'Full'], x=df.value_counts('tier', sort=False), autopct='%1.1f%%')
plt.show()
```

```python
plt.figure(figsize=(5,8))
plt.title("Hospital location")
plt.pie(labels=['Urban', 'Rural'], x=df.value_counts('location', sort=False), autopct='%1.1f%%')
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (6,8))

ax = sns.boxplot(x="size", y="beds", data=df, showfliers=True)
ax = sns.swarmplot(x="size", y="beds", data=df, color="0.25")

plt.show()
```

```python
fig, ax = plt.subplots(figsize = (6,8))

ax = sns.boxplot(x="tier", y="beds", data=df, showfliers=True)
ax = sns.swarmplot(x="tier", y="beds", data=df, color="0.25")

plt.show()
```
