```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme()

df = pd.read_csv('data/hospitals.csv')

df["size"] = df["size"].astype("category")
df["size"] = df["size"].cat.set_categories(['Small', 'Medium', 'Large'], ordered=True)
df["tier"] = df["tier"].astype("category")
df["tier"] = df["tier"].cat.set_categories(['Basic', 'Extended', 'Full'], ordered=True)
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize = (12,5))
fig.suptitle('Distribution of hospitals')

ax1.set_title('Hospital sizes')
ax1.pie(labels=['Small < 250 beds', 'Medium', 'Large > 750 beds'], x=df.value_counts('size', sort=False), autopct='%1.1f%%')

ax2.set_title("Hospital tiers")
ax2.pie(labels=['Basic', 'Extended', 'Full'], x=df.value_counts('tier', sort=False), autopct='%1.1f%%')

ax3.set_title("Hospital location")
ax3.pie(labels=['Urban', 'Rural'], x=df.value_counts('location', sort=False), autopct='%1.1f%%')

plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize = (12,8))
fig.suptitle('Hospital beds and sizes/ tiers')

sns.boxplot(ax=axes[0], x="size", y="beds", data=df, showfliers=True)
sns.swarmplot(ax=axes[0], x="size", y="beds", data=df, color="0", size=8)
axes[0].set(title='Beds per Size', xlabel='')

sns.boxplot(ax=axes[1], x="tier", y="beds", data=df, showfliers=True)
sns.swarmplot(ax=axes[1], x="tier", y="beds", data=df, color="0", size=8)
axes[1].set(title='Beds per Tier', xlabel='')

plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize = (12,8), sharey="all")
fig.suptitle('Distribution by areas')

df.value_counts('supply_area').plot(ax=ax1, kind='bar')
ax1.set(title='By supply areas', xlabel='')

df.value_counts('dispatch_area').plot(ax=ax2, kind='bar')
ax2.set(title='By dispatch area', xlabel='');
```
